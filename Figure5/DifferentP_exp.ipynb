```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots 

plt.figure(figsize=(11, 6))
plt.style.use(['ieee'])
plt.rcParams['font.size'] = '14'
linew = 3
plt.grid()
plt.tight_layout()

with open('DifferentP_04exp2sqrt2_n=300_numIter=49_k=2.352709_lambda=9.txt', 'r') as file:
    exp2sqrt2Lines = np.array([line.split() for line in file.readlines()[1::2]]).astype('int')

with open('DifferentP_04exp4_n=300_numIter=49_k=2.352709_lambda=9.txt', 'r') as file:
    exp4Lines = np.array([line.split() for line in file.readlines()[1::2]]).astype('int')

with open('DifferentP_04exp4sqrt2_n=300_numIter=49_k=2.352709_lambda=9.txt', 'r') as file:
    exp4sqrt2Lines = np.array([line.split() for line in file.readlines()[1::2]]).astype('int')

with open('DifferentP_04exp8_n=300_numIter=49_k=2.352709_lambda=9.txt', 'r') as file:
    exp8Lines = np.array([line.split() for line in file.readlines()[1::2]]).astype('int')

with open('DifferentP_04exp8_n=300_numIter=49_k=2.352709_lambda=9.txt', 'r') as file:
    pVal = np.array(file.readlines()[0::2]).astype('float').squeeze()

xVal = [1/x for x in pVal]

exp2sqrt2Lines = exp2sqrt2Lines*pVal[:,np.newaxis]
exp4Lines = exp4Lines*pVal[:,np.newaxis]
exp4sqrt2Lines = exp4sqrt2Lines*pVal[:,np.newaxis]
exp8Lines = exp8Lines*pVal[:,np.newaxis]

exp2sqrt2Lines = exp2sqrt2Lines*np.exp(-0.4* (2**1.5))
exp4Lines = exp4Lines*np.exp(-0.4* (2**2))
exp4sqrt2Lines = exp4sqrt2Lines*np.exp(-0.4* (2**2.5))
exp8Lines = exp8Lines*np.exp(-0.4* (2**3))

plt.errorbar(xVal, np.mean(exp8Lines, axis=1), yerr=np.std(exp8Lines, axis=1), label = 'Truncated $0 - 8$', marker='o', linewidth=linew, capsize=5)
plt.errorbar(xVal, np.mean(exp4sqrt2Lines, axis=1), yerr=np.std(exp4sqrt2Lines, axis=1), label = 'Truncated $0 - 4\sqrt{2}$', marker='o', linewidth=linew, capsize=5)
plt.errorbar(xVal, np.mean(exp4Lines, axis=1), yerr=np.std(exp4Lines, axis=1), label = 'Truncated $0 - 4$', marker='o', linewidth=linew, capsize=5)
plt.errorbar(xVal, np.mean(exp2sqrt2Lines, axis=1), yerr=np.std(exp2sqrt2Lines, axis=1), label = 'Truncated $0 - 2\sqrt{2}$', marker='o', linewidth=linew, capsize=5)

plt.xlabel('$1/p$')
plt.ylabel('$T \cdot p \cdot P[D \geq d]$')
plt.legend()
plt.subplots_adjust(left=0.08)
plt.savefig('DifferentP_exp.pgf', format='pgf')

plt.show()
```
