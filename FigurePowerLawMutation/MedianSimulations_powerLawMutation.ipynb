```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import scienceplots 

plt.style.use(['ieee'])
matplotlib.rcParams.update({"font.family": "serif","font.serif": ["DejaVu Serif"]})
plt.figure(figsize=(11, 6))
plt.rcParams['font.size'] = '14'
linew = 3
plt.grid()
plt.tight_layout()


n_values = np.linspace(60, 200, 15).astype(int)

with open('../FigureExpUnifDistortions/lnn15fixed_simulate_exp04_iter=49_from=60_to=200_steps=10_cutoff=1000000.txt', 'r') as file:
    data = file.readlines()
exp_plusData = np.array([data[1+4*i].split()[1:] for i in (range(len(n_values)))]).astype(int)
exp_commaData = np.array([data[2+4*i].split()[1:] for i in (range(len(n_values)))]).astype(int)

with open('lnn15fixed_simulate_power1.5_iter=49_from=60_to=200_steps=10_cutoff=1000000.txt', 'r') as file:
    data = file.readlines()
pow15_plusData = np.array([data[1+4*i].split()[1:] for i in (range(len(n_values)))]).astype(int)
pow15_commaData = np.array([data[2+4*i].split()[1:] for i in (range(len(n_values)))]).astype(int)

with open('lnn15fixed_simulate_power2.5_iter=49_from=60_to=200_steps=10_cutoff=1000000.txt', 'r') as file:
    data = file.readlines()
pow25_plusData = np.array([data[1+4*i].split()[1:] for i in (range(len(n_values)))]).astype(int)
pow25_commaData = np.array([data[2+4*i].split()[1:] for i in (range(len(n_values)))]).astype(int)

with open('lnn15fixed_simulate_power3.5_iter=49_from=60_to=200_steps=10_cutoff=1000000.txt', 'r') as file:
    data = file.readlines()
pow35_plusData = np.array([data[1+4*i].split()[1:] for i in (range(len(n_values)))]).astype(int)
pow35_commaData = np.array([data[2+4*i].split()[1:] for i in (range(len(n_values)))]).astype(int)

plt.axhline(y=1000000, linestyle='--', label='Cutoff', linewidth=linew, color='dimgrey')
plt.plot(n_values, np.median(exp_plusData, axis=1), label=r'$(1 + \lambda)$-EA with #mutations from $\mathrm{Bin}(n, 1/n)$', marker='o', linewidth=linew, color="black", linestyle = '-')
plt.plot(n_values, np.median(exp_commaData, axis=1), label=r'$(1 , \lambda)$-EA with #mutations from $\mathrm{Bin}(n, 1/n)$', marker='o', linewidth=linew, color="black", linestyle = ':')
plt.plot(n_values, np.median(pow15_plusData, axis=1), label=r'$(1 + \lambda)$-EA with #mutations from $P(k) \propto k^{-1.5}$', marker='o', linewidth=linew, color="darkorange", linestyle = '-')
plt.plot(n_values, np.median(pow15_commaData, axis=1), label=r'$(1 , \lambda)$-EA with #mutations from $P(k) \propto k^{-1.5}$', marker='o', linewidth=linew, color="darkorange", linestyle = ':')
plt.plot(n_values, np.median(pow25_plusData, axis=1), label=r'$(1 + \lambda)$-EA with #mutations from $P(k) \propto k^{-2.5}$', marker='o', linewidth=linew, color="royalblue", linestyle = '-')
plt.plot(n_values, np.median(pow25_commaData, axis=1), label=r'$(1 , \lambda)$-EA with #mutations from $P(k) \propto k^{-2.5}$', marker='o', linewidth=linew, color="royalblue", linestyle = ':')
plt.plot(n_values, np.median(pow35_plusData, axis=1), label=r'$(1 + \lambda)$-EA with #mutations from $P(k) \propto k^{-3.5}$', marker='o', linewidth=linew, color="seagreen", linestyle = '-')
plt.plot(n_values, np.median(pow35_commaData, axis=1), label=r'$(1 , \lambda)$-EA with #mutations from $P(k) \propto k^{-3.5}$', marker='o', linewidth=linew, color="seagreen", linestyle = ':')


plt.xlabel('n')
plt.ylabel('Median Number of Generations')
plt.yscale('log')
plt.legend(loc='upper right')
plt.savefig('MedianSimulations_powerLawMutation.pdf', format='pdf')

plt.show()

# rote und schware hier rein, slowdown same, wie viele mutations nicht so wichtig, schwarse schelchter weil jump to lower onemax when distortion found
```
