```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import scienceplots 

plt.style.use(['ieee'])
matplotlib.rcParams.update({"font.family": "serif","font.serif": ["DejaVu Serif"]})
plt.figure(figsize=(11, 6))
plt.rcParams['font.size'] = '14'
linew = 3
plt.grid()
plt.tight_layout()

with open('oneRun_exp04_n=150_1e9_plus.txt', 'r') as file:
    lines = file.readlines()
plus_t = [int(line.split()[0]) for line in lines] 
plus_distortion = [float(line.split()[1]) for line in lines]
plus_fitness = [float(line.split()[2]) for line in lines]
plus_onemax = [float(line.split()[2])-float(line.split()[1]) for line in lines]

with open('oneRun_exp04_n=150_1e9_comma.txt', 'r') as file:
    lines = file.readlines()
comma_t = [int(line.split()[0]) for line in lines]
comma_distortion = [float(line.split()[1]) for line in lines]
comma_fitness = [float(line.split()[2]) for line in lines]
comma_onemax = [float(line.split()[2])-float(line.split()[1]) for line in lines]

plt.axhline(y=147.88, linewidth=linew, color='black', linestyle='--', label='$n-k^*$')

plt.plot(plus_t, plus_fitness, color='royalblue', linestyle='-', linewidth=linew, label = "$(1+\lambda)$-EA Run: Total Fitness", markevery=range(len(plus_t)-1), zorder=2)
plt.plot(plus_t, plus_onemax, color='royalblue', linestyle='-.', linewidth=linew, label ="$(1+\lambda)$-EA Run: OneMax Fitness", markevery=range(len(plus_t)-1), zorder=1)
plt.plot(plus_t, plus_distortion, color='royalblue',  linestyle=':', linewidth=linew, label ="$(1+\lambda)$-EA Run: Distortion", markevery=range(len(plus_t)-1), zorder=0)

plt.plot(comma_t, comma_fitness, color='darkorange',linestyle='-', linewidth=linew, label="$(1,\lambda)$-EA Run: Total Fitness", zorder=2)
plt.plot(comma_t, comma_onemax, color='darkorange', linestyle='-.', linewidth=linew, label="$(1,\lambda)$-EA Run: OneMax Fitness", zorder=1)
plt.plot(comma_t, comma_distortion, color='darkorange', linestyle=':', linewidth=linew, label="$(1,\lambda)$-EA Run: Distortion", zorder=0)

plt.xlabel('t')
plt.xscale('log')
plt.yticks([15*i for i in range(12)])
plt.legend(loc='right')
plt.savefig('OneRun_1e9.pdf', format='pdf')
plt.show()
```
